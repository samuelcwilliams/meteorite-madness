### Light Data Wrangling

The first question that comes to mind about this dataset is how the data is spread over the years. My intuition tells me that there will be far less observations as the years go back, because older meteorites are generally discovered, rather than observed. 

Initially, I created a dataframe from the output of the `value_counts` function, which automatically counts the occurences of each value in a given column. However, this didn't deliver quite what I was looking for, `year` column served as the index rather than as an accessible column. However, this can be avoided by specifically referencing `counts.index` and `counts.values` as columns while creating the `occ` dataframe.

```python
counts = df['year'].value_counts()
occ = pd.DataFrame({'year': counts.index, 'impacts': counts.values})
occ
```

Using the `counts` dataframe, we should be able to easily visualize the impacts over time:

```python
# need to modify this so that it's easier to read. also would be nice to have a plot that goes from 1800 to 2020 so
# that more detail can be seen.
plt.bar(occ['year'], occ['impacts'])
plt.xlabel('Year');
plt.ylabel('Impacts');
plt.title('Meteorite Impacts per Year');
```
